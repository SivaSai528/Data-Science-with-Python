```python
import pandas as pd
df=pd.read_csv("10000 Sales Records.csv")
print(df["Units Sold"])
len(df.dtypes)
```

```python
df["SV"]=None
df["S"]=None
df["SsV"]=None
df["SVs"]=None
df["SVss"]
len(df.dtypes)
```

```python
d=df.dropna(axis=0,how=all,thresh=2,inplace=False)
d
```

```python
len(d.dtypes)
```

```python
import numpy as np
df["Units Sold"].replace(np.nan,999)
```

```python
df.dtypes
```

```python
df['Country'] = df['Country'].astype('string')
```

```python
df.dtypes
```

```python
#coverting all object data type into strings
obj_columns = df.select_dtypes(include=np.object).columns.tolist()
df[obj_columns] = df[obj_columns].astype('string')
#coverting all int data type into float
obj_columns = df.select_dtypes(include=np.int64).columns.tolist()
df[obj_columns] = df[obj_columns].astype('float64')
```

```python
df.dtypes
```

```python
df['Country']
```

```python
#changing data of country in lower case letters 
df['Country'] = df['Country'].str.lower()
#removing White Spaces
df['Country'] = df['Country'].str.strip()
```
