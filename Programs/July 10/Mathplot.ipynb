```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4,5],[1,2,1,3,2])
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,1,3,2])
plt.plot(x,y,color="r",linestyle="--",marker="o",markerfacecolor='white')
plt.plot([1,2,3,4],[2,2,2,2])
plt.show()
```
