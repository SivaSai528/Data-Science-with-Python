```python
import matplotlib.pyplot as plt
import numpy as np
Country=["USA","INDIA","SINGAPORE","PRANCE","UK","JAPAN","China"]
GDP=[16000,26000,19000,8000,12000,20021,20202]
plt.title("Bar Graph")
plt.xlabel("Country")
plt.ylabel("GDP per Country")
plt.bar(Country,GDP,color=["violet","indigo","blue","green","yellow","orange","red"])
plt.show()
```

```python
plt.barh(Country,GDP)
plt.title("Bar Graph")
plt.ylabel("Country")
plt.xlabel("GDP per Country")
plt.show()
```

```python
y=np.array([26,13,29])
plt.pie(y)
plt.box()
```

```python
plt.pie(y,labels=["S","A","I"])
plt.show()
```

```python
plt.pie(y,labels=["S","A","I"])
plt.legend()
plt.show()
```

```python
plt.pie(y,labels=["S","A","I"])
plt.legend(title="SAI")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.pie(y)
plt.subplot(1,2,2)
plt.bar(Country,GDP)
plt.show()
```
