```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X=np.array([df['RUNS-S']]).reshape(-1,1)
Y=df["SR-B"]
x_train,x_test,y_train,y_test=train_test_split(X,Y)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
print("Test Data Score :",model.score(x_test,y_test))
print("Train Data Score :",model.score(x_train,y_train))
```

```python
Y_pre=model.predict(x_test)
```

```python
print("Predict Test Data Score :",model.score(x_test,Y_pre))
```

```python
x_new=np.array([1775,9,1107,417,781]).reshape(-1,1)
y_new=model.predict(x_new)
```

```python
print("New Data Predict Score :",model.score(x_new,y_new))
```
